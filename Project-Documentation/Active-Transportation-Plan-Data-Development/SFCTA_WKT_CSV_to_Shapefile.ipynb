```python
import pandas as pd
import geopandas as gp
from shapely import wkt
```

```python
!ls 'County, City and Caltrans Bike Ped Network and Counts Data/SFMTA/'
```

```python
#read csv
mta_bikeway = pd.read_csv('County, City and Caltrans Bike Ped Network and Counts Data/SFMTA/MTA.bikewaynetwork.csv')
```

```python
mta_bikeway.head(5)
```

```python
#create geodataframe from csv
mta_bikeway_gdf = gp.GeoDataFrame(mta_bikeway,
                                  geometry=mta_bikeway['shape'].apply(wkt.loads)
                                  ,crs='EPSG:4326')
```

```python
mta_bikeway_gdf.to_file('County, City and Caltrans Bike Ped Network and Counts Data/SFMTA/mta_bikeway',
                        driver='ESRI Shapefile')
```
