### Merge with census geocodes to get county

```python
all_geo = pd.read_csv('Data/all-geocodes-v2016.csv',dtype={'State Code (FIPS)':str,'County Code (FIPS)':str})
```

```python
all_geo.head(5)
```

```python
cty_codes = all_geo.loc[(all_geo['Summary Level'] == 50) & (all_geo['State Code (FIPS)'] == '06')].copy()
```

```python
cty_codes.head(5)
```

```python
acs_2014_2016_2018_compare_geo = pd.merge(acs_2014_2016_2018_compare,
                                     cty_codes[['County Code (FIPS)','Area Name']],
                                     left_on='county_fips',right_on='County Code (FIPS)',
                                    how='left')
```

```python
acs_2014_2016_2018_compare_geo.rename(columns={'Area Name':'county_name',
                                               'COC_FLAG_2018':'coc_flag_acs2016',
                                               'cocflag_2013':'coc_flag_pba2035',
                                               'tot_pop_pov_det':'tot_pop_pov_det_2050',
                                               'tot_pop_civ_ni':'tot_pop_civ_ni_2050',
                                               'tot_hh':'tot_hh_2050',
                                               'tot_pop_over5':'tot_pop_over5_2050',
                                               'over75_halfsd':'over75_halfsd_2050',
                                               'minority_halfsd':'minority_halfsd_2050',
                                               'spfam_halfsd':'spfam_halfsd_2050',
                                               'disab_halfsd':'disab_halfsd_2050',
                                               'lep_halfsd':'lep_halfsd_2050',
                                               'below200_halfsd':'below200_halfsd_2050',
                                               'zvhh_halfsd':'zvhh_halfsd_2050',
                                               'hus_rent50_halfsd':'hus_rent50_halfsd_2050'
                                              },inplace=True)
```

```python
acs_2014_2016_2018_compare_geo.columns.to_list()
```

```python
cols = ['geoid',
 'state',
 'county_fips',
 'tract',
 'county_name',
 'tot_pop_2050',
 'tot_pop_pov_det_2050',
 'tot_pop_civ_ni_2050',
 'tot_hh_2050',
 'pop_zvhhs_2050',
 'tot_fam_2050',
 'tot_pop_over5_2050',
 'pop_hus_rent50_2050',
 'pop_minority_2050',
 'pop_over75_2050',
 'pop_spfam_2050',
 'pop_lep_2050',
 'pop_below200_2050',
 'pop_disability_2050',
 'pct_minority_2050',
 'pct_over75_2050',
 'pct_spfam_2050',
 'pct_lep_2050',
 'pct_below200_2050',
 'pct_disab_2050',
 'pct_zvhhs_2050',
 'pct_hus_rent50_2050',
 'over75_halfsd_2050',
 'minority_halfsd_2050',
 'spfam_halfsd_2050',
 'disab_halfsd_2050',
 'lep_halfsd_2050',
 'below200_halfsd_2050',
 'zvhh_halfsd_2050',
 'hus_rent50_halfsd_2050',
 'pct_over75_2040',
 'pct_minority_2040',
 'pct_spfam_2040',
 'pct_lep_2040',
 'pct_below200_2040',
 'pct_disab_2040',
 'pct_zvhhs_2040',
 'pct_hus_rent50_2040',
 'coc_flag_pba2035',
 'coc_flag_acs2016',
 'coc_flag_pba2040',
 'coc_flag_pba2050',
 'gain_loss_pba2040_pba2050',
 'coc_class']

coc_qa_qc = acs_2014_2016_2018_compare_geo[cols].copy()
```

```python
coc_qa_qc.head(20)
```

```python
coc_qa_qc['gain_loss_acs2016_pba2050'] = coc_qa_qc['coc_flag_pba2050'] - coc_qa_qc['coc_flag_acs2016']
```

### Merge with census tract geography

```python
state = "'06'"
counties = "('001','013','041','055','075','081','085','095','097')"
url = 'https://tigerweb.geo.census.gov/arcgis/rest/services/Generalized_ACS2018/Tracts_Blocks/MapServer/3/query?where=STATE={state}+AND+COUNTY+IN{counties}&outFields=GEOID,NAME&f=geojson'.format(state=state,counties=counties)
```

```python
geo_rq = requests.get(url)
tracts_json = geo_rq.json()
```

```python
tracts_2018_gdf = gp.GeoDataFrame.from_features(tracts_json['features'],crs=tracts_json['crs']['properties']['name'])
```

```python
coc_qa_qc_geo = pd.merge(tracts_2018_gdf,coc_qa_qc,left_on='GEOID',right_on='geoid',how='inner')
```

```python
coc_qa_qc_geo.plot(figsize=(12,12),column='coc_flag_pba2050')
```

```python
coc_qa_qc_geo.to_file("Data/COCs_ACS_2009_2014_2016_2018_geog.geojson", driver="GeoJSON")
```
